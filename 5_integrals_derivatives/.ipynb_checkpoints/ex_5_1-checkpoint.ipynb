```python
import numpy as np
```

```python
t = np.loadtxt('velocities.txt', dtype=int, usecols=0) # s
v = np.loadtxt('velocities.txt', usecols=1) # m/s
```

$$
s(t) = \int{v(t)}
$$

```python
def integrate_trapz(y, x):
    n = len(x)
    h = (x[-1] - x[0]) / n
    return h * (((y[0] + y[-1]) / 2) + sum([y[i] for i in range(1, n)]))
```

```python
s = [integrate_trapz(v[:i], t[:i]) for i in range(1, v.size)]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot(t, v, label='speed (m/s)')
ax.plot(t[1:], s, label='distance (m)')

ax.set_xlabel('time (s)')
ax.legend()
plt.show()
```
