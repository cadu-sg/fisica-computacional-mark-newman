#### a)

$$
J_m(x) = \frac{1}{\pi} \int_{0}^{\pi} cos(m\theta - x sin\theta) d\theta
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# REGRA DE SIMPSON ESTENDIDA
# --------------------------
def integrate_simps(func, a, b, n):
    h = (b - a) / n    
    odd_sum = sum([func(a + k * h) for k in range(1, n, 2)])
    even_sum = sum([func(a + k * h) for k in range(2, n, 2)])
    return (h / 3) * (func(a) + func(b) + 4 * odd_sum + 2 * even_sum)
```

```python
def J(m, x):
    integrand = lambda theta : np.cos(m * theta - x * np.sin(theta))
    return (1 / np.pi) * integrate_simps(integrand, 0, np.pi, 1000)
```

```python
J(0, 20)
```

```python
domain = np.linspace(0, 20)

J_0 = [J(0, x) for x in domain]
J_1 = [J(1, x) for x in domain]
J_2 = [J(2, x) for x in domain]
```

```python
fig, ax = plt.subplots()

ax.plot(domain, J_0, label='$J_0$')
ax.plot(domain, J_1, label='$J_1$')
ax.plot(domain, J_2, label='$J_2$')

ax.legend()
plt.show()
```

#### b)

Intensidade da luz no padrão de difração circular consistindo de um ponto central rodeado por uma série de anéis concêntricos. Esse padrão é produzido por uma luz de estrela com comprimento de onda $\lambda$ ao passar pela abertura circular de um telescópio e ser focada pelo telescópio no plano focal.

$$
I(r) = \bigg( \frac{J_1(kr)}{kr} \bigg) ^ 2
$$

- $r$ é a distância no plano focal do centro do padrão de difração
- $k = 2 \pi / \lambda$
- $J_1(x)$ é uma função é Bessel
- $\lambda$ é o comprimento de onda

```python
def I(radius, wavelength):
    k = (2 * np.pi) / wavelength
    kr = k * radius
    return (J(1, kr) / kr) ** 2
```

```python
wavelength = 5e-7   # m
radius_start = 0   # m
radius_stop = 1e-6 # m


light = []
for r in np.linspace(1e-7, 1e-6):
    light.append(I(r, wavelength))
```
